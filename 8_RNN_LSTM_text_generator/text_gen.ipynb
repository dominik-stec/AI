```
np_chars = np.empty([2])
k_list = np.empty([1], dtype='U1')
v_list = np.empty([1], dtype=np.int8)
for k in chars_idx.keys():
  k_list = np.append(k_list, k) 
  v_list = np.append(v_list, chars_idx[k])
v_list = v_list[1:]
k_list = k_list[1:]
kv_list = np.empty(0, dtype=object)
kv_list = np.append(kv_list, v_list)
kv_list = np.append(kv_list, k_list)
kv_list = np.array_split(kv_list, 2)
kv_list = np.stack((kv_list[0], kv_list[1]), axis=-1) 
kv_list
```

Metoda one hot encoding


```
one_hot = OneHotEncoder()
one_hot_v = one_hot.fit_transform(kv_list).toarray()
one_hot_v = np.array(one_hot_v, dtype=np.int0)
print("liczba liter")
len(one_hot_v)
```

```
def one_hot_cod_char(char):
  idx = chars_idx[char]
  return one_hot_v[idx]
print('przykład kodowania one hot')
one_hot_cod_char('\n')
```

Metoda kodująca one hot dla ciągu znaków

```
def one_hot_code_text(text, start_idx, stop_idx):
  chunk = text[start_idx: stop_idx]
  ret = np.empty(0, dtype=np.int0)
  for i in range(len(chunk)):
    c = chunk[i]
    c_hot = one_hot_cod_char(c)
    ret = np.append(ret, c_hot)
  ret = np.array_split(ret, len(chunk))
  ret = np.array(ret)
  return ret
print('5 znaków, 174 wartości dla kodowania jednego znaku')
one_hot_code_text(book, 5, 10).shape
```

Budowanie modelu.

Warstwa wejściowa stanowi zbiór znaków przetwarzanego tekstu.

Warstwy ukryte stanowią neurony LSTM (ang. Long Short-Term Memory)(długa pamięć krótkoterminowa). Reprezentują rekurencyjne głębokie sieci neuronowe z pamięcią. W przeciwieństwie do tradycyjnych RNN, LSTM nie są podatne na zanikanie sygnału uczącego podczas propagacji synału w głąb sieci, dzięki czemu mogą przetwarzać długie dane wejściowe bez zaniku procesu uczenia wraz ze wzrostem głębokości przetwarzania danych w sieci.

źródło:
https://en.wikipedia.org/wiki/Long_short-term_memory

Użycie LSTM w projekcie zapewnia generowanie pasujących wg zadanego modelu znaków, użytych do tworzenia słów w zdaniach. Dzięki zastosowanemu podejściu, podstawą przetwarzania danych, może być wyuczenia na podstawie bardzo długich ciągów znaków ułożonych wg spójnego schematu jak np w dziele literackim.

Metoda definiuje:

ilość użytych znaków, liczbę warstw LSTM, liczbę neuronów LSTM, parametr dropout,

zwraca w wyniku gotowy model.

Użycie klasy opakowującej, (wrappera) TimeDistrubuted, umożliwia generowanie ciągu znaków, na który składają się litery, odstępy (spacje i białe znaki), budujące całe zdania, bazujac na podziale na kroki czasowe (timestamps). Dla każdego kroku czasowego generowany jest jeden znak, na podstawie pobranych z wcześniejszych wartw LSTM danych.

Warstwa LSTM odpowiada za dobór odpowiednich znaków, wrapper TimeDistrubuted łączy znaki w całe zdania.

funkcja aktywacji: softmax

(oblicza prawdopobieństwa użycia poszczególnych znaków w kroku czasowym)

funkcja straty: categorical_crossentropy 

(wybiera kategorię znaku o największej wartości prawdopodobieństwa, poprzez minimalizację wartości zoptymalizowanych wag neuronów sieci)

optymalizator:

RMSprop

(aktualizuje wagi neuronów, w celu optymalizacji funkcji straty)

źródło:

https://ichi.pro/pl/przeglad-roznych-optymalizatorow-dla-sieci-neuronowych-247308806799555

parametry optymalizacyjne:

dropout: 0.1 (domyśnie)

(liczba losowo pomijanych neuronów w iteracji)

learning rate: 0.01

(efektywność uczenia w iteracjach, tempo uczenia)

```
def get_model(num_chars, num_layers, num_nodes, dropout = 0.1):
  input = Input(shape=(None, num_chars), name='input')
  first_layer = input
  for i in range(num_layers):
      lstm = LSTM(num_nodes, return_sequences=True, name=f'lstm_hidden_layer_{i+1}')(first_layer)
      if dropout:
        first_layer = Dropout(dropout)(lstm)
      else:
        first_layer = lstm
  #dense_output = TimeDistributed(Dense(num_chars, name='dense', activation='softmax'))(first_layer)
  dense_output = Dense(num_chars, name='dense', activation='softmax')(first_layer)
  model = Model(inputs=[input], outputs=[dense_output])
  optimizer = RMSprop(lr=0.01)
  model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
  return model
```

Generator losowych fragmentów utworu.

parametry:

text - całość tekstu

char_idx - zindeksowany słownik znaków występujących w tekście

num_of_words - liczba generowanych słów

num_of_letters - liczba liter w słowie

Metoda jest generatorem zmapowanych na binarne indeksy znaków pobranych z wylosowanego fragmentu tekstu o podanej długości i o podanej liczbie znaków.

Generator zwiększa losowość danych użytych do trenowania.

```
def words_bin_gen(text, char_idx, num_of_words, num_of_letters):
  X = np.zeros([num_of_words, num_of_letters, len_chars], dtype=np.int8)
  y = np.zeros([num_of_words, num_of_letters, len_chars], dtype=np.int8)
  while True:
    for row in range(num_of_words):
      id = random.randrange(len_text - num_of_letters - 1)
      words_bin_idx = np.zeros((num_of_letters + 1, len_chars))
      for i in range(num_of_letters + 1):
        rand_word_from_text = text[id+1]
        char_idx = chars_idx[rand_word_from_text]
        words_bin_idx[i, char_idx] = 1
      X[row,:,:] = words_bin_idx[:num_of_letters]
      y[row,:,:] = words_bin_idx[1:]
      #print(y[0][0])
    yield X, y
```
